```python
%%writefile main_hybrid_pipeline.py
import argparse
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
from src.data_loader import JSONLLoader
from src.evaluation import AdvancedEvaluationSuite
from src.utils import extract_primary_tag


def main():
    parser = argparse.ArgumentParser(
        description="Run hybrid pipeline using offline Task 1 predictions + online Task 2 QA Model"
    )
    parser.add_argument("--val_path", type=str, required=True)
    parser.add_argument("--t1_pred_path", type=str, required=True)
    parser.add_argument("--t2_checkpoint", type=str, required=True)
    parser.add_argument(
        "--output_path", type=str, default="final_complete_submission.jsonl"
    )
    args = parser.parse_args()

    # 1. Ingest evaluation assets
    val_df = JSONLLoader(args.val_path).load_data()
    t1_df = pd.read_json(args.t1_pred_path, lines=True)

    # 2. Reconstruct Task 1 tags by mapping UUIDs dynamically
    pred_map = dict(zip(t1_df["uuid"], t1_df["spoilerType"]))
    pred_tags = [pred_map.get(uid, "phrase") for uid in val_df["uuid"]]

    # 3. Load Task 2 Extractive QA Transformer weights
    print(f"[INFO] Loading trained Task 2 QA Checkpoint from {args.t2_checkpoint}...")
    tokenizer = AutoTokenizer.from_pretrained(args.t2_checkpoint)
    model = AutoModelForQuestionAnswering.from_pretrained(args.t2_checkpoint)
    model.eval()

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # 4. Generate text spoilers via token-span extraction
    print("Running extractive inference on validation partition...")
    predicted_spoilers = []

    for idx, row in val_df.iterrows():
        question = (
            " ".join(row["postText"])
            if isinstance(row["postText"], list)
            else str(row["postText"])
        )
        context = (
            " ".join(row["targetParagraphs"])
            if isinstance(row["targetParagraphs"], list)
            else str(row["targetParagraphs"])
        )

        inputs = tokenizer(
            question,
            context,
            truncation="only_second",
            max_length=512,
            return_tensors="pt",
        )
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model(**inputs)

        start_idx = int(np.argmax(outputs.start_logits.cpu().numpy()))
        end_idx = int(np.argmax(outputs.end_logits.cpu().numpy()))

        input_ids = inputs["input_ids"][0].cpu().numpy()
        if end_idx >= start_idx:
            extracted_tokens = input_ids[start_idx : end_idx + 1]
            pred_text = tokenizer.decode(
                extracted_tokens, skip_special_tokens=True
            ).strip()
        else:
            pred_text = ""

        # Fallback heuristic optimization if the model extracts an empty string
        if not pred_text and len(row["targetParagraphs"]) > 0:
            pred_text = row["targetParagraphs"][0]

        predicted_spoilers.append([pred_text] if pred_text else [])

    # 5. Compute full dissertation evaluation suite metrics
    y_true_tags = val_df["tags"].apply(extract_primary_tag).tolist()
    ground_truth_spoilers = val_df["spoiler"].tolist()

    print("\n=== FINAL INTEGRATED NEURAL SYSTEM REPORT ===")
    AdvancedEvaluationSuite.evaluate_task1(y_true_tags, pred_tags)
    AdvancedEvaluationSuite.evaluate_task2(ground_truth_spoilers, predicted_spoilers)

    # 6. Export verified submission artifact
    output_records = [
        {
            "uuid": r["uuid"],
            "spoilerType": pred_tags[i],
            "spoiler": predicted_spoilers[i],
        }
        for i, r in val_df.iterrows()
    ]
    pd.DataFrame(output_records).to_json(args.output_path, orient="records", lines=True)
    print(
        f"\n[SUCCESS] Integrated neural prediction file saved safely to {args.output_path}"
    )


if __name__ == "__main__":
    main()
```
